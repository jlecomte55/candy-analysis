### OneHotEncoder

One-hot encode all previously indexed categorical features.

```
oneHotEnc = OneHotEncoder()
oneHotEnc.setInputCols(["cat_competitorname", "cat_state", "cat_region", "cat_division"])
oneHotEnc.setOutputCols(["vec_competitorname", "vec_state", "vec_region", "vec_division"])
```

Test your solution:

```
assert set(oneHotEnc.getInputCols()) == set(["cat_competitorname", "cat_state","cat_region", "cat_division"]), 'oneHotEnc expected inputCols: "cat_competitorname", "cat_state", "cat_pounds", "cat_places", "cat_region", "cat_division"'
        
assert set(oneHotEnc.getOutputCols()) == set(["vec_competitorname", "vec_state", "vec_region", "vec_division"]), 'oneHotEnc expected outputCols: "vec_competitorname", "vec_state", "vec_pounds", "vec_places", "vec_region", "vec_division"'

print("Tests passed.")
```
