```
gold_candy.registerTempTable("Gold_Candy")
```

```
regions = spark.sql("select distinct(Region) from Gold_Candy").rdd.map(lambda row : row[0]).collect()
regions.sort()
```

```
dbutils.widgets.dropdown("regions", "Northeast", [str(x) for x in regions])
```

```
%sql
select 
distinct(competitorname) as Candy_Name,
SUM(POUNDS) as Pounds
from Gold_Candy
where region = getArgument("regions")
GROUP BY Pounds, Candy_Name
ORDER BY Pounds desc
```

```
%sql
select 
STATE as State,
SUM(POUNDS) as Pounds
from Gold_Candy
where region = getArgument("regions")
GROUP BY Pounds, State
ORDER BY Pounds desc
```

```
states = spark.sql("select distinct(STATE) from Gold_Candy").rdd.map(lambda row : row[0]).collect()
states.sort()
```

```
dbutils.widgets.dropdown("states", "VA", [str(x) for x in states])
```

```
%sql
select 
competitorname as Candy_Name,
SUM(POUNDS) as Pounds
from Gold_Candy
where State = getArgument("states")
GROUP BY Pounds, Candy_Name
ORDER BY Pounds desc
```

```
%sql
select 
competitorname as Candy_Name,
winpercent as Win_Percent,
STATE as State,
POUNDS as Pounds
from Gold_Candy
where region = getArgument("regions")
ORDER BY Pounds desc
```
