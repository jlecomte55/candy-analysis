### Training a Decision Tree Regressor

Let's take a look at the expert parameters of Decision Trees

```
from pyspark.ml.regression import DecisionTreeRegressor

dt = DecisionTreeRegressor()

print(dt.explainParams())
```

Now, train a Decision Tree Regressor model:

```
display(full_predictedDF)
```

```
dt_candy_features = full_predictedDF.drop('prediction','STATE','Region','Division','vec_competitorname','vec_state','vec_region','vec_division','features','cat_state')
display(dt_candy_features)
```

```
print(featureCols)
```

```
featureCols = [c for c in dt_candy_features.columns if c != "POUNDS" if c != "competitorname"]
assembler = VectorAssembler(inputCols=featureCols, outputCol="features")
vecDF = assembler.transform(dt_candy_features)

(trainDF, testDF) = vecDF.randomSplit([0.95, 0.05], seed=273)

(dt.setLabelCol("POUNDS")
   .setMaxDepth(7)
)
dtModel = dt.fit(trainDF)
```

Let's evaluate our Decision Tree Regressor Model:

```
predictedDF = dtModel.transform(testDF)

rmse = evaluator.setMetricName("rmse").evaluate(predictedDF)
mae = evaluator.setMetricName("mae").evaluate(predictedDF)
r2 = evaluator.setMetricName("r2").evaluate(predictedDF)
print("RMSE: {}\nMAE: {}\nR2: {}".format(rmse, mae, r2))
display(predictedDF)
```

What are the features with the most predictive power in this dataset?

```
imp = dtModel.featureImportances
colNames = dict(zip(range(0, len(featureCols)), featureCols))
importances = sorted(zip(imp.values, imp.indices), reverse=True)
print("Importance of features:\n")

for f in importances:
  print((colNames[f[1]] + " (id=" + str(f[1]) + "): ").ljust(27) + str(f[0]))
```
