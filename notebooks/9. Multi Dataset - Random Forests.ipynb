### Training a Random Forest Regressor

Let's take a look at the expert parameters of the Random Forest Regressor

```
from pyspark.ml.regression import RandomForestRegressor

rf = RandomForestRegressor()

print(rf.explainParams())
```

Now, train a Random Forest Regressor model:

```
(rf.setLabelCol("POUNDS")
   .setMaxDepth(6)
   .setNumTrees(50)
)
rfModel = rf.fit(trainDF)
```

Let's evaluate our Random Forest Regressor Model:

```
predictedDF = rfModel.transform(testDF)

rmse = evaluator.setMetricName("rmse").evaluate(predictedDF)
mae = evaluator.setMetricName("mae").evaluate(predictedDF)
r2 = evaluator.setMetricName("r2").evaluate(predictedDF)
print("RMSE: {}\nMAE: {}\nR2: {}".format(rmse, mae, r2))
display(predictedDF)
```

#### What are the features with the most predictive power in each tree?

```
print("Number of trees trained: {}\nFeature importances in individual trees:\n".format(len(rfModel.trees)))
for ind, dtModel in enumerate(rfModel.trees):
  print("Decision Tree #{}".format(ind + 1))
  imp = dtModel.featureImportances
  colNames = dict(zip(range(0, len(featureCols)), featureCols))
  importances = sorted(zip(imp.values, imp.indices), reverse=True)
  for f in importances:
    print(("  " + colNames[f[1]] + " (id=" + str(f[1]) + "): ").ljust(30) + str(f[0]))
  print()
```
